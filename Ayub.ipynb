##### Define a custom function (called 'clean_text') that performs the following operations on a text: (a) change all characters to lowercase, and remove any leading and/or trailing spaces, (b) adjust string.punctuation (if necessary), and use it to remove all punctuation characters, (c) remove all numbers, (d) remove all NLTK stopwords, and (e) split the text in a list of words. Then use this function to clean the first text and to split it in a list of words.

```python
def remove_punctions(text):
    p = '''!()-[]{};:'"\,<>./?@#$%^&*_~â€”‘“’'''
    for ele in text:
        if ele in p:
            text = text.replace(ele, "")
    return text
def clean_text(text):
    text = text.lower()
    text = text.strip()
    text = text.replace(" ", "")
    text_cleaned = remove_punctions(text)
    text_cleaned_digits = ''.join([i for i in text_cleaned if not i.isdigit()])
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text_cleaned_digits )
    filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]
    filtered_sentence = []
    for w in word_tokens:
        if w not in stop_words:
            filtered_sentence.append(w)
    return filtered_sentence,text_cleaned_digits
cleaned_text_18 ,text_cleaned_str = clean_text(DataFrame_18)
print(cleaned_text_18)
```

##### Calculate the frequency of all words in the first (cleaned) text, save the words and counts as key-value pairs in a dictionary, and print the dictionary.

```python
def wordFreq(cleaned_text):
    wordfreq = []
    word_freq = {}
    for w in cleaned_text:
        word_freq[w] = cleaned_text.count(w)
    return word_freq
word_freq = wordFreq(cleaned_text_18)
print(word_freq)
```

##### Print the 25 words (and their counts) with the highest frequency in the first text. Print each of these words on a separate line and in the following order: first the word, then its count.

```python
from operator import itemgetter
def get_popular_words(word_freq):
    res = dict(sorted(word_freq.items(), key = itemgetter(1), reverse = True)[:25])
    for word, count in res.items():
        print("Word: ", word, " Count: ", count)
```

###### Use the Python codes that have been discussed earlier in the course to perform a similarity analysis based on the three texts. That is, perform the following operations in three subsequent cells: (a) Define a 'custom_tokenizer' function that uses NLTK's word tokenizer and the Porter stemmer, (b) Use Scikit-learn's CountVectorizer class to convert the three texts to bag-of-words vectors, and (c) Use Scikit-learn to calculate the cosine similarity measures. (Hint: Keep in mind that you have to perform operation (b) based on the cleaned texts; i.e., not on the uncleaned texts or on a list of words from the cleaned texts.)

```python
def custom_tokenizer(data):
    ps = PorterStemmer()
    stemmed = ps.stem(data)
    word_tokens = word_tokenize(stemmed)
    return word_tokens
```

```python
documents = [cleaned_text_18 , cleaned_text_19 , cleaned_text_mst]
document = [' '.join(str(item)) for item in documents ]
count_vectorizer = CountVectorizer(tokenizer=custom_tokenizer)
count_vecs = count_vectorizer.fit_transform(document )
print(count_vectorizer.get_feature_names()[:10])
print(count_vecs.toarray()[:, :10])
```

```python
cosine_sim_matrix = cosine_similarity(count_vecs)
print(cosine_sim_matrix)
```
